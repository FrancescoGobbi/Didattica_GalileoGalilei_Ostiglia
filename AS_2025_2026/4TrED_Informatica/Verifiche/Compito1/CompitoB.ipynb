```python
# Esercizio 3:
# Creare un programma in Python che prenda in input 5 valori da parte dell’utente. Il programma
# deve eseguire e stampare la somma e la media per i soli valori dispari tra quelli inseriti.
# ESEMPIO: Se n1 = 5, n2 = 10, n3 = 2, n4 = 9 ed n5 = 1. La somma sarà essere 19, in quanto solo 5, 9
# e 5 sono numeri dispari. Di conseguenza la media sarà 6,3333 .

# Inserimento dei valori interi
x1 = int(input("Inserisci un valore: "))
x2 = int(input("Inserisci un valore: "))
x3 = int(input("Inserisci un valore: "))
x4 = int(input("Inserisci un valore: "))
x5 = int(input("Inserisci un valore: "))

# Inizializzo somma e contatore
somma = 0
count = 0

# Controllo se i numeri inseriti sono dispari
if (x1 % 2 == 1): # Controllo se è dispari il numero
    print("Il numero", x1, "è dispari")
    somma = somma + x1
    count = count + 1 # count+=1
if (x2 % 2 == 1): # Controllo se è dispari il numero
    print("Il numero", x2, "è dispari")
    somma = somma + x2
    count = count + 1 # count+=1
if (x3 % 2 == 1): # Controllo se è dispari il numero
    print("Il numero", x3, "è dispari")
    somma = somma + x3
    count = count + 1 # count+=1
if (x4 % 2 == 1): # Controllo se è dispari il numero
    print("Il numero", x4, "è dispari")
    somma = somma + x4
    count = count + 1 # count+=1
if (x5 % 2 == 1): # Controllo se è dispari il numero
    print("Il numero", x5, "è dispari")
    somma = somma + x5
    count = count + 1 # count+=1

# Eseguo la media
if (count > 0):
    media = somma / count
    print("La somma è", somma, "mentre la media è", media)
else:
    print("La somma è", somma, "mentre la media è 0")
```

```python
# Esercizio 4:
# Creare un programma in Python che prenda in input un valore n (da definire il tipo di oggetto).
# Il programma deve eseguire il prodotto dei primi n valori.
# ESEMPIO: Se n = 5, il programma dovrà stampare 15, in quanto 5*4*3*2*1 = 120

# Inserimento del valore intero positivo
n = int(input("Inserisci un valore intero positivo: "))

# Inizializzo il prodotto
prodotto = 1

# Eseguo il prodotto dei primi n numeri
for i in range(1, n + 1):
    prodotto = prodotto * i

# Stampo il risultato
print("Il prodotto dei primi", n, "numeri è:", prodotto)
```

```python
# Esercizio 5:
# Creare un programma in Python che prenda in input un numero (a) e stampi i primi a-numeri
# primi possibili.
# ESEMPIO: Se a = 5 si devono stampare i primi 5 numeri primi, quindi: 2, 3, 5, 7 e 11

# Input del numero di numeri primi da stampare
a = int(input("Inserisci quanti numeri primi vuoi stampare: "))

count = 0  # Contatore dei numeri primi trovati
num = 2    # Numero da testare se è primo, ovviamente si parte dal primo numero primo
divisori = 0 # Contatore dei divisori

# Ciclo fino a quando non ho trovato 'a' numeri primi
while count < a:
    divisori = 0 # Reset del contatore dei divisori per il nuovo numero
    for i in range(1, num + 1): # Controllo i divisori del numero
        if num % i == 0: # Se 'i' è un divisore di 'num'
            divisori += 1 # Incremento il contatore dei divisori
    if (divisori == 2): # Controllo se il numero di divisori è 2 (1 e se stesso)
        print(num) # Stampo il numero primo trovato
        count += 1 # Ho trovato un numero primo, incremento il contatore
    num += 1 # Passo al numero successivo
```
