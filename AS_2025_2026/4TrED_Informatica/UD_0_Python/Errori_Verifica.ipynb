```python
# Esempio di doppi cicli for, quindi cicli annidati, uno dentro l'altro
# N.B. L'oggetto di iterazione deve essere esclusivo di tale iterazione
# Ogni iterazione deve avere il suo oggetto di iterazione
# Questo per non avere sia errori nella scrittura del codice, per migliorare la leggibilità del codice
# e per evitare errori di logica nel codice

n = 5
m = 3 

for i in range(1, n+1): # Vado da 1 a 5
    for j in range(1, m+1): # Vado 1 a 3
        print(f"({i},{j})")
```

```python
# Esempio di ciclo for semplice, in cui si legge un numero 5 di valori da input e si stampa
# L'oggetto di iterazione non viene usato all'interno del ciclo, quindi si usa il carattere underscore _

n = 5
for _ in range(1, n+1): # Vado da 1 a 5
    x = int(input(" "))
    print(x)
```

```python
# Esempio di ciclo for semplice, in cui si stampa "Ciao" per 5* 2 volte, in quanto si hanno due cicli annidati
# L'oggetto di iterazione non viene usato all'interno del ciclo, quindi si usa il carattere underscore _
# POSSIBILE ERRORE DI LOGICA: utilizzare _ come variabile di iterazione in entrambi i cicli annidati
# potrebbe creare confusione, per quanto in Python questo possa funzionare correttamente
# Si consiglia di usare nomi diversi per le variabili di iterazione nei cicli annidati per migliorare la leggibilità del codice

n = 5
m = 2
for _ in range(1, n+1): # Vado da 1 a 5
    for _ in range(1, m+1): # Vado da 1 a 2
        print("Ciao")
```

```python
# Esempio di doppi cicli for, quindi cicli annidati, uno dentro l'altro
# Stampa tutte le coppie (i,j) con 1 <= i <= n e i <= j <= n
# Nota: il secondo ciclo inizia da i, non da 1
# Questo significa che per ogni valore di i, j parte da i e va fino a n
# Di conseguenza, il numero di stampe per ogni i diminisce all'aumentare di i
# Esempio quindi di cicli annidati in cui l'oggetto di iterazione del secondo ciclo dipende dal primo ciclo

n = 5
m = 2
for i in range(1, n+1): # Viene eseguita 5 volte 
    for j in range(i, n+1): # 
        print(f"({i},{j})")
```
