```python
# Esesercizio 1 - Crezione di una classe

# Creare una classe in Python che rappresenti una scuola, quindi ogni oggetto della
# rappresenta uno specifico studente.
# Ogni studente deve poter vedere la sua media dei voti, aggiungere un voto e calcolare la nuova media.
# Ongi studente deve anche essere in grado di poter visualizzare il proprio nome e cognome.
# 

# ESEMPIO: se lo studente "Mario Rossi" ha i voti [6, 7, 8], la sua media è 7.0.

class Studente:
    def __init__(self, nome, cognome):
        self.nome = nome
        self.cognome = cognome
        self.voti = []

    def aggiungi_voto(self, voto):
        self.voti.append(voto)

    def calcola_media(self):
        if len(self.voti) == 0:
            return 0
        return sum(self.voti) / len(self.voti)

    def visualizza_nome_cognome(self):
        return f"{self.nome} {self.cognome}"
    
# Esempio di utilizzo
studente1 = Studente("Mario", "Rossi")
studente1.aggiungi_voto(6)
studente1.aggiungi_voto(7)
studente1.aggiungi_voto(8) 
print(studente1.visualizza_nome_cognome())  # Output: Mario Rossi
print(studente1.calcola_media())  # Output: 7.0
```

```python
# Esercizio 2 - Creazione di funzioni in Python

# Creare una funzione in Python che prenda in input due numeri interi a e b
# e restituisca la somma di tutti i numeri compresi tra a e b, estremi inclusi.

# ESEMPIO (a=3, b=7) --> 3+4+5+6+7 = 25
# ESEMPIO (a=10, b=4) --> 4+5+6+7+8+9+10 = 54

def somma_intervallo(a, b):
    if a > b:
        max = a
        min = b
    else:
        max = b
        min = a
    somma = 0
    for i in range(min, max + 1):
        somma += i
    return somma

# Esempio di utilizzo
print(somma_intervallo(3, 7))  # Output: 25
print(somma_intervallo(10, 4))  # Output: 54
```

```python
# Esercizio 3 - Creazione di funzione in Python

# Creare una funzione in Python che prenda in input un numero intero n ed un numero m.
# La funzione deve conteggiare quante coppie sono possibili di numeri interi (a, b) tali che:
# 1 <= a < b <= n e a + b sia multiplo di m.

# ESEMPIO (n=7, m=3) --> Le coppie sono (1,2), (1,5), (2,4), (3,6), (4,5) --> Risultato = 5
# ESEMPIO (n=5, m=4) --> Le coppie sono (1,3), (2,6) --> Risultato = 2

def conta_coppie(n, m):
    count = 0
    for a in range(1, n):
        for b in range(a + 1, n + 1):
            if (a + b) % m == 0:
                count += 1
    return count

# Esempio di utilizzo
print(conta_coppie(7, 3))  # Output: 5
print(conta_coppie(5, 4))  # Output: 2
```

```python
# Esercizio 4 - Creazione di una funzione in Python

# Creare una funzione in Python che prenda in input un numero intero n.
# La funzione deve verificare se il numero n, considerato come un anno, è bisestile o no.
# La funzione deve restituire True se l'anno è bisestile, False altrimenti.

# ESEMPIO (n=2020) --> True
# ESEMPIO (n=2021) --> False
# ESEMPIO (n=1200) --> True

def is_bisestile(n):
    if (n % 4 == 0 and n % 100 != 0) or (n % 400 == 0):
        return True
    return False

# Esempio di utilizzo
print(is_bisestile(2020))  # Output: True
print(is_bisestile(2021))  # Output: False
print(is_bisestile(1200))  # Output: True
```
