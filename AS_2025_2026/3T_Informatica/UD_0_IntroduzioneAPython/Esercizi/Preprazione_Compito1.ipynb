```python
# Eserecizio 2:
# Crea un codice in Python che crei tre oggetti interi scelti da voi (Es A, B, C).
# Il codice deve eseguire il prodotto tra i tre oggetti creati in precedenza.
# Successivamente deve sottrarre al prodotto calcolato il valore puntato da A.
# Successivamente deve aggiungere al valore puntato da B il valore di C.

# Stapare il risultato ed il tipo di ogni oggetto creato e puntato.

# Creo gli oggetti e le etichette a cui puntano gli oggetti creati da voi utenti programmatiori
A = 10
B = 20
C = 30

# Eseguo il prodotto tra i tre oggetti (creo un nuovo oggetto per il risultato del prodotto)
prodotto = A * B * C

# Sottraggo al prodotto il valore puntato da A (creo un nuovo oggetto per il risultato della sottrazione, facendolo puntare sempre all'etichetta prodotto)
prodotto = prodotto - A
# Aggiungo al valore puntato da B il valore di C (creo un nuovo oggetto per il risultato della somma, facendolo puntare sempre all'etichetta B) 
B = B + C

# Stampo il risultato ed il tipo di ogni oggetto creato e puntato
print("Il valore di A è: ", A)
print("Il tipo di A è: ", type(A))

print("Il valore di B è: ", B)
print("Il tipo di B è: ", type(B))

print("Il valore di C è: ", C)
print("Il tipo di C è: ", type(C))

print("Il valore di prodotto è: ", prodotto)
print("Il tipo di prodotto è: ", type(prodotto))
```

```python
# Esercizio 3:
# Crea un codice in Python che crei tre oggetti float scelti da voi (Es D, E, F).
# Il codice deve eseguire la somma tra i tre oggetti creati in precedenza.
# Successivamente deve moltiplicare per 2 il valore puntato da D.
# Successivamente deve dividere per 3 il valore puntato da F.
# Stapare il risultato, quanti byte occupano in memoria ed il tipo di ogni oggetto creato e puntato.

import sys

# Creo gli oggetti e le etichette a cui puntano gli oggetti creati da voi utenti programmatiori
D = 1.5
E = 2.5
F = 3.5

# Eseguo la somma tra i tre oggetti (creo un nuovo oggetto per il risultato della somma)
somma = D + E + F
# Moltiplico per 2 il valore puntato da D (creo un nuovo oggetto per il risultato della moltiplicazione, facendolo puntare sempre all'etichetta D)
D = D * 2
# Divido per 3 il valore puntato da F (creo un nuovo oggetto per il risultato della divisione, facendolo puntare sempre all'etichetta F)
F = F / 3

# Stampo il risultato, quanti byte occupano in memoria ed il tipo di ogni oggetto creato e puntato
print("Il valore di D è: ", D)
print("Il tipo di D è: ", type(D))
print("I byte occupati in memoria da D sono: ", sys.getsizeof(D))

print("Il valore di E è: ", E)
print("Il tipo di E è: ", type(E))
print("I byte occupati in memoria da E sono: ", sys.getsizeof(E))

print("Il valore di F è: ", F)
print("Il tipo di F è: ", type(F))
print("I byte occupati in memoria da F sono: ", sys.getsizeof(F))

print("Il valore di somma è: ", somma)
print("Il tipo di somma è: ", type(somma))
print("I byte occupati in memoria da somma sono: ", sys.getsizeof(somma))
```
