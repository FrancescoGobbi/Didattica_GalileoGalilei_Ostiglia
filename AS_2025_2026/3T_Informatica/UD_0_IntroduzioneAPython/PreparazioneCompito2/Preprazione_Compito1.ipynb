```python
# Esercizio 1:
# Creare un codice in Python che, dato in input un numero intero N,
# calcoli se N è un numero primo o no.
# Come si fa a considerare se un numero è primo oppure no?
# Il programma deve stampare "Primo" se N è un numero primo, "Non Primo" altrimenti.

# ESEMPIO (N=7) --> Primo
# ESEMPIO (N=8) --> Non Primo
# ESEMPIO (N=13) --> Primo

n = int(input("Inserisci un numero intero N: ")) # Prendo in input un numero intero N

count = 0 # Inizializzo il contatore a zero

for i in range(1, n+1): # Ciclo da 1 a N
    if n % i == 0: # Controllo se N è divisibile per i
        count += 1 # Incremento il contatore

if count == 2: # Se il contatore è uguale a 2, allora N è primo
    print(f"{n} è un numero primo.")
else:
    print(f"{n} non è un numero primo.")
```

```python
# Esercizio 3:
# Creare un codice in Python che, dato in input un numero intero N ed un numero intero M,
# conti quanti numeri sono primi tra N ed M, estremi inclusi.
# Il programma deve stampare il conteggio dei numeri primi trovati.

# ESEMPIO (N=3, M=7) --> 4 (3, 5, 7 sono primi)
# ESEMPIO (N=10, M=15) --> 2 (11, 13 sono primi)

n = int(input("Inserisci un numero intero N: ")) # Prendo in input un numero intero N
m = int(input("Inserisci un numero intero M: ")) # Prendo in input un numero intero M

count_primi = 0 # Inizializzo il contatore dei numeri primi a zero

# Determino il min e il max tra n e m, in quanto l'utente può inserire i numeri in qualsiasi ordine
if n > m:
    max = n
    min = m
else:
    max = m
    min = n

for i in range(min, max+1): # Ciclo da min a max
    count = 0 # Inizializzo il contatore a zero
    for j in range(1, i+1): # Ciclo da 1 a i, per controllare se i è primo
        if i % j == 0: # Controllo se i è divisibile per j
            count += 1 # Incremento il contatore
    if count == 2: # Se il contatore è uguale a 2, allora i è primo
        count_primi += 1 # Incremento il contatore dei numeri primi
    
print(f"Tra {n} e {m} ci sono {count_primi} numeri primi.") # Stampo il conteggio dei numeri primi trovati
```
