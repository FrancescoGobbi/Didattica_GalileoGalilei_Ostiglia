```python
# Esercizio 1:
# Creare un codice in Python che, dato in input un numero intero N,
# calcoli se N è un numero primo o no.
# Come si fa a considerare se un numero è primo oppure no?
# Il programma deve stampare "Primo" se N è un numero primo, "Non Primo" altrimenti.

# ESEMPIO (N=7) --> Primo
# ESEMPIO (N=8) --> Non Primo
# ESEMPIO (N=13) --> Primo

n = int(input("Inserisci un numero intero N: ")) # Prendo in input un numero intero N

count = 0 # Inizializzo il contatore a zero

for i in range(1, n+1): # Ciclo da 1 a N
    if n % i == 0: # Controllo se N è divisibile per i
        count += 1 # Incremento il contatore

if count == 2: # Se il contatore è uguale a 2, allora N è primo
    print(f"{n} è un numero primo.")
else:
    print(f"{n} non è un numero primo.")
```
