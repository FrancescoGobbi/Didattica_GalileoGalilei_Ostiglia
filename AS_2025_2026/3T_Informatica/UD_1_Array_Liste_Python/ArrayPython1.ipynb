```python
# ESERCIZIO 1

# Scrivere un programma in Python che prenda in input un numero intero n.
# Il programma deve chiedere in input n numeri interi ed inserirli in un array.
# Il programma deve trovare e stampare il valore massimo e minimo presente nell'array.

# ESEMPIO:
# Se n = 5 e i numeri inseriti sono: 12, 7, 22, 5, 15
# Il programma deve stampare:
# Valore massimo: 22
# Valore minimo: 5

# TO DO...

# Importare la libreria ArrayLib
import array as arr # as serve per abbreviare il nome della libreria

# Prendo in input la dimensione del vettore
n = int(input("Inserisci la dimensione del vettore: "))

# Creo il vettore di interi vuoto
my_array = arr.array('i') # Creo un array di interi vuoto, qunindi senza valori iniziali

# Prendo in input n-valori interi utilizzando un for
for i in range(n): # for i in range(0,n):
    val = int(input(f"Inserisci un valore nel vettore in posizione {i}: "))
    my_array.append(val) # Aggiungo/appendo il numero inserito (val) nel vettore
# Fine del ciclo for

print(my_array) # Stampa diretta e totale del vettore

# INIZIO LA SOLUZIONE DELL'ESERCIZIO: TROVO IL MASSIMO ED IL MINIMO DEL VETTORE
# N.B. Inizializzo i valori di massimo e minimo utilizzando il nome dell'etichetta per l'oggetto NON come max o min
# In quanto max e min sono già dei nomi di funzioni predefinite in Python, quindi è meglio evitare di usarli come nomi di variabili per non creare confusione
# Altrimenti, Python andrà a sostituire l'alis max e min con i valori di massimo e minimo, e non potremo più utilizzare le funzioni max() e min() per trovare il massimo e il minimo in una sola riga di codice
# Al riavvio dell'ambiente Python le funzioni max() e min() torneranno a funzionare correttamente, in quanto le variabili max e min non esisteranno più
maxVal = my_array[0] # Inizializzo il massimo al primo elemento del vettore
minVal = my_array[0] # Inizializzo il minimo al primo elemento del vettore

# Vado a ricercare il massimo ed il minimo nel vettore
for i in range(0,n): # Così abbiamo gli indici di posizione che vanno da 0 a n-1
    # Controllo per il massimo
    if (my_array[i] > maxVal):
        maxVal = my_array[i]
        
    # Controllo per il minimo
    if (my_array[i] < minVal):
        minVal = my_array[i]

print(f"Il valore massimo è: {maxVal}")
print(f"Il valore minimo è: {minVal}")

# Trovo il massimo in una sola riga di codice
max2 = max(my_array) # La funzione max() restituisce il valore massimo presente nel vettore my_array
print(f"Il valore massimo è: {max2}")   

# Trovo il minimo in una sola riga di codice
min2 = min(my_array) # La funzione min() restituisce il valore minimo presente nel vettore my_array
print(f"Il valore minimo è: {min2}")   
```

```python
# ESERCIZIO 2
# Scrivere un programma in Python che prenda in input un numero intero n ed un intero m.
# Il programma deve chiedere in input n numeri interi ed inserirli in un array.
# Il programma deve calcolare e stampare la somma e la media dei numeri maggiori di m presenti nell'array.

# ESEMPIO:
# Se n = 6, m = 10 e i numeri inseriti sono: 8, 15, 22, 5, 10, 30
# Il programma deve stampare:
# Somma dei numeri maggiori di 10: 67
# Media dei numeri maggiori di 10: 22.33  

# TO DO...

# Importare la libreria ArrayLib
import array as arr # as serve per abbreviare il nome della libreria

# Prendo in input la dimensione del vettore
n = int(input("Inserisci la dimensione del vettore: "))
# Prendo in input il valore di m per l'esercizio
m = int(input("Inserisci il valore m: "))

# Creo il vettore di interi vuoto
my_array = arr.array('i') # Creo un array di interi vuoto, qunindi senza valori iniziali

# Prendo in input n-valori interi utilizzando un for
for i in range(n): # for i in range(0,n):
    val = int(input(f"Inserisci un valore nel vettore in posizione {i}: "))
    my_array.append(val) # Aggiungo/appendo il numero inserito (val) nel vettore
# Fine del ciclo for

# Creo degli oggetti per la soluzione
somma = 0 # La somma è intera
media = 0.0 # Perché vogliamo la media float
count = 0 # Questo sarà il denomiatore per la mia media

# Soluzione dell'esercizio
for i in range(0,n): # for i in range(0,n):
    if (my_array[i] > m):
        somma = somma + my_array[i]
        count = count + 1
    # Fine if
# Fine for

media = somma / count

print(f"La somma è: {somma}")
print(f"La media è: {media}")

# SOLUZIONE ALTERNATIVA
for i in my_array: # Vado a scorrere direttamente gli elementi del vettore, senza utilizzare gli indici di posizione
    # In questo caso i rappresenta direttamente l'elemento del vettore, quindi non devo scrivere my_array[i] ma solo i
    # Automaticamente, il ciclo for scorrerà tutti gli elementi del vettore my_array, assegnando ad i ogni elemento in sequenza
    if (i > m):
        somma = somma + i
        count = count + 1
    # Fine if
# Fine for

media = somma / count

print(f"La somma è: {somma}")
print(f"La media è: {media}")
```

```python
# ESERCIZIO 6
# Scrivere un programma in Python che prenda in input un numero intero n.
# Il programma deve chiedere in input n numeri interi ed inserirli in un array.
# Il programma deve ordinare l'array in ordine crescente e stampare l'array ordinato.

# ESEMPIO:
# Se n = 5 e i numeri inseriti sono: 12, 7, 22, 5, 15
# Il programma deve stampare:
# Array ordinato in ordine crescente: [5, 7, 12, 15, 22]

# TO DO...

# Importare la libreria ArrayLib
import array as arr # as serve per abbreviare il nome della libreria

# Prendo in input la dimensione del vettore
n = int(input("Inserisci la dimensione del vettore: "))
# Creo il vettore di interi vuoto
my_array = arr.array('i') # Creo un array di interi vuoto, quindi senza valori iniziali

# Prendo in input n-valori interi utilizzando un for
for i in range(n): # for i in range(0,n):
    val = int(input(f"Inserisci un valore nel vettore in posizione {i}: "))
    my_array.append(val) # Aggiungo/appendo il numero inserito (val) nel vettore
# Fine del ciclo for   

print(f"Array iniziale: {my_array}")

# Soluzione dell'esercizio utilizzando una doppia iterazione (Bubble Sort)
for i in range(len(my_array)):
    for j in range(0, len(my_array) - 1):
        if my_array[j] > my_array[j + 1]:
            tmp = my_array[j]
            my_array[j] = my_array[j + 1]
            my_array[j + 1] = tmp

# Fine iterazione: array ordinato in ordine crescente

print(f"Array ordinato in ordine crescente: {my_array}")
```

```python
# ESERCIZIO 9
# Scrivere un programma in Python che prenda in input un numero intero n.
# Il programma deve chiedere in input n numeri interi ed inserirli in un array.
# Il programma deve calcolare la media di tutti i numeri primi presenti nel vettore.

# ESEMPIO:
# Se n = 6 e i numeri inseriti sono: 8, 15, 22, 5, 10, 7
# I numeri primi nel vettore sono due, ovvero 5 e 7.
# Il programma deve quindi eseguire la media tra 5 e 7, ovvero 6.00
# (5+7) / 2 = 6.00

# TO DO...

# Importare la libreria ArrayLib
import array as arr # as serve per abbreviare il nome della libreria  

# Prendo in input la dimensione del vettore
n = int(input("Inserisci la dimensione del vettore: "))
# Creo il vettore di interi vuoto
my_array = arr.array('i') # Creo un array di interi vuoto, quindi senza valori iniziali

# Prendo in input n-valori interi utilizzando un for
for i in range(n): # for i in range(0,n):
    val = int(input(f"Inserisci un valore nel vettore in posizione {i}: "))
    my_array.append(val) # Aggiungo/appendo il numero inserito (val) nel vettore
# Fine del ciclo for    

# Soluzione dell'esercizio
sommaPrimi = 0 # La somma dei numeri primi è inizializzata a 0, in quanto la somma di 0 numeri è 0
countPrimi = 0 # Il contatore dei numeri primi è inizializzato a 0, in quanto non abbiamo ancora contato nessun numero primo
for i in range(n): # for i in range(0,n):
    countDivisori = 0 # Azzero il contatore dei divisori per ogni numero del vettore
    for j in range(1, my_array[i]+1): # Controllo se il numero è primo, quindi controllo se è divisibile per qualche numero tra 2 e il numero stesso - 1
        if (my_array[i] % j == 0): # Se il numero è divisibile per j, allora non è primo
            countDivisori = countDivisori + 1 # Incremento il contatore dei divisori
        # Fine if
    # Fine for

    if (countDivisori == 2): # Se il numero è divisibile solo per 1 e per se stesso, allora è primo
        sommaPrimi = sommaPrimi + my_array[i] # Aggiungo il numero primo alla somma dei numeri primi
        countPrimi = countPrimi + 1 # Incremento il contatore dei numeri primi
    # Fine if
# Fine for

# Calcolo la media dei numeri primi
mediaPrimi = sommaPrimi / countPrimi
print(f"La media dei numeri primi è: {mediaPrimi}")
```
