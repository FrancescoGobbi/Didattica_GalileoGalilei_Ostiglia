## 1. Come viene passato un vettore a una funzione

Quando passiamo un array a una funzione, nella funzione **non viene copiato tutto il vettore** in automatico come in una descrizione teorica semplificata: la funzione lavora sul riferimento all'oggetto.

Dal punto di vista pratico, per il nostro corso possiamo dire che:

- la funzione **riceve il vettore** come parametro;
- può **leggere** i suoi valori;
- in alcuni casi può anche **modificarli**.

Struttura generale:

```python
def nome_funzione(vettore):
    # istruzioni
```

Dove:

- `nome_funzione` è il nome della funzione;
- `vettore` è il parametro che rappresenta l'array passato alla funzione.

### Valori necessari

Per lavorare con un vettore, spesso servono:

- il **vettore stesso**;
- la **lunghezza** del vettore, ottenibile con `len(vettore)`;
- eventualmente un valore aggiuntivo, per esempio il numero di elementi da inserire.


````python
# N.B. Le funzioni con il passaggio di array ricevono il riferimento 
# di memoria dell'array, quindi modificando l'array all'interno della funzione
# si modifica anche l'array originale passato come argomento.

import array as arr

def funzione(vettore):
    print(id(vettore)) # Stampo l'indirizzo di memoria del vettore ricevuto
    vettore[0] = 10 # Modifico il primo elemento del vettore
    # Quindi dell'oggetto puntato, ovvero del vettore originale

if __name__ == "__main__":
    vettore = arr.array('i', [1, 2, 3, 4, 5]) # Creo un vettore di interi
    print(id(vettore)) # Stampo l'indirizzo di memoria del vettore originale
    print("Prima della funzione:", vettore)
    funzione(vettore)
    print("Dopo la funzione:", vettore)
````

## 2. Funzione per calcolare la media dei valori del vettore

### Spiegazione

La media aritmetica si calcola sommando tutti i valori e dividendo il risultato per il numero totale degli elementi.

Formula:

**media = somma dei valori / numero dei valori**

Esempio matematico:

Se il vettore contiene i valori `2, 4, 6, 8`, allora:

- somma = 2 + 4 + 6 + 8 = 20
- numero di elementi = 4
- media = 20 / 4 = 5

La funzione riceve un array come parametro e restituisce la media dei suoi elementi.


````python
import array as arr

def media_vettore(vettore):
    # Controlliamo che il vettore non sia vuoto
    if len(vettore) == 0:
        return 0

    # Variabile per accumulare la somma degli elementi
    somma = 0

    # Scorriamo tutti i valori del vettore
    for valore in vettore:
        somma = somma + valore

    # Calcoliamo e restituiamo la media
    return somma / len(vettore)


# Main()
v1 = arr.array('i', [2, 4, 6, 8])
print("Vettore:", v1)
print("Media:", media_vettore(v1))
````

## 3. Funzione per ordinare un vettore

### Spiegazione

Ordinare un vettore significa disporre i suoi elementi in un certo ordine, di solito:

- **crescente**: dal più piccolo al più grande;
- **decrescente**: dal più grande al più piccolo.

Esempio matematico:

Se il vettore è:

`[7, 3, 9, 1]`

in ordine crescente diventa:

`[1, 3, 7, 9]`

Nel seguente esempio realizziamo una funzione che ordina il vettore in modo crescente.

Per semplicità utilizziamo il metodo del **confronto tra elementi**.


````python
import array as arr

def ordina_vettore(vettore):

    # Algoritmo di ordinamento Bubble Sort
    for i in range(len(vettore)):
        for j in range(i + 1, len(vettore)):
            if vettore[i] > vettore[j]:
                # Scambio dei due valori
                temp = vettore[i]
                vettore[i] = vettore[j]
                vettore[j] = temp

    # Restituiamo il vettore ordinato
    return vettore


# Main()
v2 = arr.array('i', [7, 3, 9, 1, 5])
print("Vettore originale:", v2)
print("Vettore ordinato:", ordina_vettore(v2))
````

## 4. Funzione per inserire in un vettore i numeri della serie di Fibonacci

### Richiamo teorico

La successione di Fibonacci è una successione numerica in cui:

- il primo termine è 0;
- il secondo termine è 1;
- ogni termine successivo si ottiene sommando i due precedenti.

I primi termini sono:

`0, 1, 1, 2, 3, 5, 8, 13, 21, ...`

Definizione matematica:

- F(0) = 0
- F(1) = 1
- F(n) = F(n-1) + F(n-2), per n >= 2

### Esempio

- F(0) = 0
- F(1) = 1
- F(2) = 1
- F(3) = 2
- F(4) = 3
- F(5) = 5

In questo esercizio realizziamo una funzione che, dato un numero `n`, costruisce un vettore contenente i primi `n` numeri della successione di Fibonacci.


````python
import array as arr

def vettore_fibonacci(n):
    # Creiamo un array vuoto di interi
    vettore = arr.array('i')

    # Gestiamo il caso n <= 0
    if n <= 0:
        return vettore

    # Inseriamo il primo numero di Fibonacci
    vettore.append(0)

    # Se n è almeno 2, inseriamo anche il secondo numero
    if n > 1:
        vettore.append(1)

    # Calcoliamo gli altri termini
    for i in range(2, n):
        nuovo_valore = vettore[i - 1] + vettore[i - 2]
        vettore.append(nuovo_valore)

    # Restituiamo il vettore completo
    return vettore


# Main()
n = 10
vf = vettore_fibonacci(n)
print(f"I primi {n} numeri della serie di Fibonacci sono:")
print(vf)
````

## 6. Esempio di funzione con modifica del vettore

Funzione che prende in input il puntatore di un vettore e modifica ogni parametro sostituendolo con la somma di tutti i valori precedenti ad esso.



````python
# ESEMPIO: Se il vettore iniziale è [1, 2, 3, 4, 5], 
# dopo la funzione diventa [1, 3, 6, 10, 15] perché:
# - Il primo elemento rimane 1
# - Il secondo elemento diventa 1 + 2 = 3
# - Il terzo elemento diventa 3 + 3 = 6
# - Il quarto elemento diventa 6 + 4 = 10
# - Il quinto elemento diventa 10 + 5 = 15

def funzione(vettore):
    for i in range(1, len(vettore)):
        vettore[i] = vettore[i] + vettore[i - 1]

if __name__ == "__main__":
    vettore = arr.array('i', [1, 2, 3, 4, 5]) # Creo un vettore di interi
    print("Prima della funzione:", vettore)
    funzione(vettore)
    print("Dopo la funzione:", vettore)
````

## 7. Altri esempi di funzione con vettori (Array in Python)

````python
# Esercizio 1
# Funzione in Python che riceve in input un vettore di interi e modifica il vettore in modo che ogni elemento
# diventi il fattoriale di se stesso. 
# Il fattoriale di un numero n è il prodotto di tutti i numeri interi da 1 a n.

def vettore_fattoriale(vettore):
    for i in range(len(vettore)): # Scorro tutti gli elementi del vettore
        fattoriale = 1 # Inizializzo la variabile per il fattoriale
        for j in range(1, vettore[i] + 1): # Calcolo il fattoriale del valore corrente
            fattoriale *= j
        vettore[i] = fattoriale # Modifico il vettore con il fattoriale calcolato

if __name__ == "__main__":
    vettore = arr.array('i', [1, 2, 3, 4, 5]) # Creo un vettore di interi
    print("Prima della funzione:", vettore)
    vettore_fattoriale(vettore) # Funzione che modifica il vettore con i fattoriali
    print("Dopo la funzione:", vettore)
````

````python
# Esercizio 2
# Funzione in Python che riceve in input un vettore di interi ed un oggetto intero n.
# La funzione restituisce il numero di occorrenze di n all'interno del vettore.

def conta_occorrenze(vettore, n):
    count = 0 # Inizializzo il contatore
    for valore in vettore: # Scorro tutti i valori del vettore
        if valore == n: # Se il valore è uguale a n
            count += 1 # Incremento il contatore
    return count # Restituisco il numero di occorrenze

if __name__ == "__main__":
    vettore = arr.array('i', [1, 2, 3, 2, 4, 2, 5]) # Creo un vettore di interi
    n = 2 # Numero da contare
    print("Vettore:", vettore)
    conta = conta_occorrenze(vettore, n) # Chiamo la funzione per contare le occorrenze ed attendo il return
    print(f"Il numero {n} appare {conta} volte nel vettore.")
````

````python
# Esercizio 3
# Funzione in Python che riceve in input un vettore di interi ed un oggetto intero n.
# La funzione modifica il vettore inserendo n al posto di tutti gli elementi che sono multipli di n.

def sostituisci_multipli(vettore, n):
    for i in range(len(vettore)): # Scorro tutti gli elementi del vettore
        if vettore[i] % n == 0: # Se l'elemento è un multiplo di n
            vettore[i] = n # Sostituisco l'elemento con n

if __name__ == "__main__":
    vettore = arr.array('i', [1, 2, 3, 4, 5, 6]) # Creo un vettore di interi
    n = 2 # Numero da sostituire
    print("Prima della funzione:", vettore)
    sostituisci_multipli(vettore, n) # Funzione che modifica il vettore sostituendo i multipli di n con n
    print("Dopo la funzione:", vettore)
````

````python
# Esercizio 4
# Funzione in Python che riceve in input un vettore di interi ed un oggetto intero n.
# La funzione restituisce un NUOVO vettore che contiene solo gli elementi del vettore originale che
# sono maggiori di n.

def filtra_maggiori(vettore, n):
    nuovo_vettore = arr.array('i') # Creo un nuovo vettore vuoto di interi
    for valore in vettore: # Scorro tutti i valori del vettore originale
        if valore > n: # Se il valore è maggiore di n
            nuovo_vettore.append(valore) # Aggiungo il valore al nuovo vettore
    return nuovo_vettore # Restituisco il nuovo vettore filtrato, passando il nuovo riferimento di memoria

if __name__ == "__main__":
    vettore = arr.array('i', [1, 5, 3, 7, 2]) # Creo un vettore di interi
    n = 3 # Numero da confrontare
    print("Vettore originale:", vettore)
    nuovo_vettore = filtra_maggiori(vettore, n) # Chiamo la funzione per filtrare i valori maggiori di n 
    # ed attendo il return, ovvero il nuovo riferimento di memoria del nuovo vettore
    print(f"Vettore con solo i valori maggiori di {n}:", nuovo_vettore)
````
