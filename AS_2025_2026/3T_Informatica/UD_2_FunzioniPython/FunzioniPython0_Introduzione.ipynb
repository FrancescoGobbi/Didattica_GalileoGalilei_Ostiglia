In questo notebook introduciamo un nuovo argomento fondamentale della programmazione: **le funzioni in Python**.

Le funzioni servono per:
- organizzare meglio il codice;
- evitare ripetizioni inutili nel codice;
- rendere i programmi più leggibili;
- suddividere un problema in parti più semplici.

Una funzione è quindi un **blocco di istruzioni** a cui viene assegnato un nome, e che può essere eseguito quando necessario.

## Significato di funzione

Una **funzione** è una porzione di codice che svolge un compito specifico.

Quindi la **funzione**:
- si **definisce** una volta;
- si può **richiamare** tutte le volte che serve;
- può avere o meno dati/oggetti in ingresso.

Pensiamo a una funzione come a una *scatola nera*: che prende in input dei dati, li elabora e di successivamente può restituire altri dati

## Struttura generale di una funzione

La struttura generale di una funzione in Python è la seguente:

```python
def nome_funzione(parametro/i):
    # istruzioni della funzione

# Inizio del programma con la funzione main()
# Richiamo della funzione
nome_funzione(...)
```

Osserviamo i vari elementi:
- `def` è la parola chiave usata per definire una funzione;
- `nome_funzione` è il nome scelto dal programmatore;
- tra parentesi si possono inserire eventuali **parametri**;
- i due punti `:` indicano l'inizio del blocco della funzione;
- il corpo della funzione deve essere **indentato**;
- per eseguire la funzione bisogna **richiamarla** usando il suo nome.

## Il concetto di parametri

I **parametri** sono informazioni che passiamo a una funzione affinché possa lavorare su dati diversi.

Per esempio, una funzione può:
- non ricevere nessun parametro;
- ricevere uno o più parametri.

Quando passiamo un valore a una funzione, quel valore viene associato al parametro corrispondente.

Esempio generale:

```python
def saluta(nome):
    print("Sono nella funzione saluta()")
    print("Ciao", nome)
    print("Fine codice funzione saluta()")

# main()
if __name__ == "__main__":
    print("Sono nella funzione main()")
    saluta("Marco")
    print("Sono nella funzione main()")
    saluta("Giulia")
    print("Sono nella funzione main()")
```

In questo caso:
- `nome` è il **parametro** della funzione;
- `"Marco"` e `"Giulia"` sono i **valori passati** alla funzione durante il richiamo.

La stessa funzione può quindi essere riutilizzata con dati diversi.

## Prima funzione senza parametri

Creiamo ora una prima funzione molto semplice, che **non riceve nessun parametro** e stampa il messaggio `"Ciao Mondo"`.

```python
# Questa funzione non ha il return
def ciao_mondo():
    print("Ciao Mondo")

# Richiamo della funzione
# Questo è il main del programma, ovvero la funzione main()
# che viene eseguita quando lanciamo il programma
ciao_mondo()
```

```python
# A livello formale in Python non esiste il concetto di main(), come in altri linguaggi
# Qualsiasi riga di codice che non è indentata (ovvero che non fa parte di una funzione) 
# viene eseguita quando lanciamo il programma

# Nella buona programmazione è utilizzzare tale dicitura
if __name__ == "__main__":
    ciao_mondo() # Questo chiamerà la funzione ciao_mondo() quando lanciamo il programma
    # funzione del blocco di codice precedente
```

In questo esempio:
- la funzione si chiama `ciao_mondo`;
- tra parentesi non c'è nulla, quindi **non richiede parametri**;
- quando viene richiamata, stampa sempre lo stesso messaggio.

## Prima funzione con un parametro

Creiamo adesso una funzione che riceve **un valore** e stampa proprio quel valore.

```python
# Questa funzione non ha il return
def stampa_valore(valore):
    print(valore)

# Richiamo della funzione con un valore
stampa_valore(10)
stampa_valore("Python")
stampa_valore(3.14)
```

In questo caso:
- `valore` è il parametro della funzione;
- possiamo passare numeri, testo o altri dati;
- la funzione stampa ogni volta il valore ricevuto.

**La funzione è sempre la stessa, ma il comportamento concreto dipende dal dato/oggetto che riceve in ingresso**.

```python
# Funzione con return
def somma(a, b):
    return a + b

if __name__ == "__main__":
    risultato = somma(5, 7)
    print("La somma è:", risultato)

    # Possiamo anche stampare direttamente il risultato della funzione senza salvarlo in una variabile
    print("La somma è:", somma(4, 7))

    risultato = somma(3.5, 2.5)
    print("La somma è:", risultato)

    print("La somma è:", somma(1.5, 2.5))
```
