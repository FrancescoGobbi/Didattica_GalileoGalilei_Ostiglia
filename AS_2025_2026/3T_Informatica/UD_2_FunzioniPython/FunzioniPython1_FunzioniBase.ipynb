## 1. Funzione per la somma tra due numeri

Creare una funzione in Python che prenda in input due numeri `a` e `b` e restituisca il valore della somma:

\[
a + b
\]

### Esempio matematico
Se `a = 7` e `b = 5`, allora:

\[

\]

La funzione riceverà due parametri in input e restituirà il risultato della somma.


```python
# Funzione che calcola la somma tra due numeri
def somma(a, b):
    # Restituisce la somma dei due parametri
    return a + b

# Esempio di utilizzo
risultato = somma(7, 5)
print("La somma è:", risultato)
```

## 3. Funzione per la potenza tra due numeri

La **potenza** indica quante volte una base viene moltiplicata per sé stessa.

Data una base `a` e un esponente `b`, vogliamo calcolare:

\[
a^b
\]

### Esempio matematico
Se `a = 2` e `b = 5`, allora:

\[
2^5 = 2 * 2 * 2 * 2 * 2 = 32
\]

In Python possiamo usare l'operatore `**` per calcolare una potenza.


```python
# Funzione che calcola la potenza tra due numeri
def potenza(base, esponente):
    # Restituisce base elevata a esponente
    return base ** esponente

# Esempio di utilizzo
risultato = potenza(2, 5)
print("La potenza è:", risultato)
```

## 4. Funzione che stampa i numeri da 1 a `n`

In questo esercizio vogliamo creare una funzione che, dato un numero intero positivo `n`, stampi tutti i numeri compresi tra `1` e `n`.

### Esempio matematico
Se `n = 5`, i numeri stampati saranno:

\[
1, 2, 3, 4, 5
\]

Questo esercizio è utile per ripassare l'uso dei **cicli** e dei **parametri** di una funzione.


```python
# Funzione che stampa i numeri da 1 a n
def stampa_da_1_a_n(n):
    # Il ciclo parte da 1 e arriva fino a n compreso
    for i in range(1, n + 1):
        print(i)

# Esempio di utilizzo
print("Numeri da 1 a 5:")
stampa_da_1_a_n(5)
```

## 5. Funzione per il fattoriale di un numero

Il **fattoriale** di un numero naturale `n` si indica con `n!` ed è definito come il prodotto di tutti i numeri interi positivi da `1` a `n`.

### Definizione teorica
Possiamo scriverlo così, in modo chiaro e leggibile:

- `n! = 1 · 2 · 3 · ... · n`
- per convenzione: `0! = 1`

In altre parole:
- se `n = 1`, allora `1! = 1`
- se `n = 2`, allora `2! = 1 · 2 = 2`
- se `n = 3`, allora `3! = 1 · 2 · 3 = 6`

### Esempio matematico
Se `n = 5`, allora:

`5! = 1 · 2 · 3 · 4 · 5 = 120`

Il fattoriale viene usato molto in matematica, ad esempio nel calcolo di disposizioni, permutazioni e combinazioni.


```python
# Funzione che calcola il fattoriale di un numero n
def fattoriale(n):
    # Caso particolare: per definizione 0! = 1
    if n == 0:
        return 1

    # Variabile che conterrà il prodotto progressivo
    risultato = 1

    # Moltiplico tutti i numeri da 1 a n
    for i in range(1, n + 1):
        risultato = risultato * i

    # Restituisco il fattoriale calcolato
    return risultato

# Esempio di utilizzo
print("Il fattoriale di 5 è:", fattoriale(5))
```

## 6. Funzione per il numero di Fibonacci in posizione `n`

La **successione di Fibonacci** è una famosa sequenza di numeri in cui ogni termine, a partire dal terzo, è dato dalla somma dei due precedenti.

### Definizione teorica
I primi termini della successione sono:

`0, 1, 1, 2, 3, 5, 8, 13, 21, ...`

La definizione matematica si può scrivere così:

- `F(0) = 0`
- `F(1) = 1`
- `F(n) = F(n-1) + F(n-2)` per `n ≥ 2`

Questo significa che:
- il valore in posizione `0` è `0`
- il valore in posizione `1` è `1`
- da quel momento in poi, ogni nuovo termine si ottiene sommando i due precedenti

### Esempio matematico
Per esempio:

- `F(2) = F(1) + F(0) = 1 + 0 = 1`
- `F(3) = F(2) + F(1) = 1 + 1 = 2`
- `F(4) = F(3) + F(2) = 2 + 1 = 3`
- `F(5) = F(4) + F(3) = 3 + 2 = 5`
- `F(6) = 8`

In questo esercizio realizziamo una funzione che restituisce il valore di Fibonacci nella posizione `n`.


```python
# Funzione che calcola il numero di Fibonacci in posizione n
def fibonacci(n):
    # Se n è 0, restituiamo 0
    if n == 0:
        return 0

    # Se n è 1, restituiamo 1
    if n == 1:
        return 1

    # Variabili per memorizzare i due termini precedenti
    precedente = 0
    corrente = 1

    # Calcolo iterativo dei termini successivi fino alla posizione n
    for i in range(2, n + 1):
        successivo = precedente + corrente
        precedente = corrente
        corrente = successivo

    # Alla fine corrente contiene F(n)
    return corrente

# Esempio di utilizzo con il main()
print("Il numero di Fibonacci in posizione 6 è:", fibonacci(6))
```

## 7. Funzione che calcola la media di `n` numeri inseriti dall'utente

In questo esercizio vogliamo creare una funzione che:
1. riceva in input un numero `n`;
2. chieda all'utente di inserire `n` numeri interi;
3. calcoli la **media aritmetica** dei valori inseriti.

### Formula matematica della media
Se abbiamo `n` valori:

\[
x_1, x_2, x_3, ..., x_n
\]

la media aritmetica è:

`media` = (x_1 + x_2 + x_3 + ... + x_n)/(n)

### Esempio matematico
Se i numeri inseriti sono `4`, `6`, `8`, allora:

`media` = (4 + 6 + 8)/(3) = 18/3 = 6



```python
# Funzione che chiede all'utente n numeri interi e ne calcola la media
def media_n_numeri(n):
    # Variabile per accumulare la somma dei numeri inseriti
    somma_valori = 0

    # Chiedo all'utente di inserire n numeri
    for i in range(1, n + 1):
        numero = int(input(f"Inserisci il numero intero {i}: "))
        somma_valori = somma_valori + numero

    # Calcolo la media dividendo la somma per il numero totale di valori
    media = somma_valori / n

    # Restituisco la media calcolata
    return media

# Esempio di utilizzo con il main() che chiede all'utente quanti numeri vuole inserire e poi stampa la media
n = int(input("Quanti numeri vuoi inserire? "))
print("La media è:", media_n_numeri(n))
```
